```python
def build_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='relu', input_shape=[len(x_train.keys())]),
    tf.keras.layers.Dense(1, activation='relu'),
    tf.keras.layers.Dense(1)
  ])


  optimizer = tf.keras.optimizers.SGD(learning_rate = 0.01)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
```

```python
model = build_model()
model.summary()

EPOCHS = 100
early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)
history = model.fit(
  x_train, y_train,
  epochs=EPOCHS, validation_data = (x_validate, y_validate), verbose=2,
  callbacks=([early_stopping]))
```
