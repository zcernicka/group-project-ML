```python
import pandas as pd
#load merged.csv
merged = pd.read_csv(r'C:\Users\zuzan\OneDrive\Desktop\Python Load\Buurt_.csv', sep = ',')
merged.buurtcode = merged.buurtcode.astype(str)
merged = merged[['buurtcode', 'buurtname']]
merged.drop_duplicates(subset ='buurtcode', keep = 'first', inplace = True)
print(merged.shape)
print(merged.head())
```

```python
import pandas as pd
school = pd.read_csv(r'C:\Users\zuzan\OneDrive\Desktop\cbs_school (3).csv', 
                     error_bad_lines = False, sep = ';')
school = school[['WijkenEnBuurten', '0-4age_Binnen5Km_55','4+age_Binnen10Km_67','4+age_Binnen10Km_71','4+age_Binnen10Km_75']]
school.rename(columns={"WijkenEnBuurten": "NBH_code"})
school.head()
```

```python
#cleaned_density['No_residents']=cleaned_density['No_residents'].str.extract(r'(\d)')
school['0-4age_Binnen5Km_55']=school['0-4age_Binnen5Km_55'].str.extract(r'(\d)')
school['4+age_Binnen10Km_67']=school['4+age_Binnen10Km_67'].str.extract(r'(\d)')
school['4+age_Binnen10Km_71']=school['4+age_Binnen10Km_71'].str.extract(r'(\d)')
school['4+age_Binnen10Km_75']=school['4+age_Binnen10Km_75'].str.extract(r'(\d)')
school
```

```python
school['0-4age_Binnen5Km_55']=school['0-4age_Binnen5Km_55'].astype(float)
school['4+age_Binnen10Km_67']=school['4+age_Binnen10Km_67'].astype(float)
school['4+age_Binnen10Km_71']=school['4+age_Binnen10Km_71'].astype(float)
school['4+age_Binnen10Km_75']=school['4+age_Binnen10Km_75'].astype(float)
```

```python
school.rename(columns={"WijkenEnBuurten": "NBH_code"},inplace = True)
school['4+age_Binnen10Km']= school[['4+age_Binnen10Km_67','4+age_Binnen10Km_71','4+age_Binnen10Km_75']].sum(axis=1)
school.head()
```

```python
school=school.drop(['4+age_Binnen10Km_67','4+age_Binnen10Km_71','4+age_Binnen10Km_75'],axis=1)
```

```python
school = school[school['NBH_code'].str.contains('BU')]
school['NBH_code'] = school['NBH_code'].astype(str)
for a in range(len(school)):
    if school['NBH_code'].iloc[a][2:3] == '0' and school['NBH_code'].iloc[a][3:4] != '0':
        school['NBH_code'].iloc[a] = ''.join(school['NBH_code'].iloc[a].rsplit('BU0'))
    elif school['NBH_code'].iloc[a][2:4] == '00' and school['NBH_code'].iloc[a][4:5] != '0':
        school['NBH_code'].iloc[a] = ''.join(school['NBH_code'].iloc[a].rsplit('BU00'))
    elif school['NBH_code'].iloc[a][2:5] == '000':
        school['NBH_code'].iloc[a] = ''.join(school['NBH_code'].iloc[a].rsplit('BU000'))
    else:
        school['NBH_code'].iloc[a] = school['NBH_code'].iloc[a][2:]
   
school.head(100)
```

```python
cleaned_school = pd.merge(school, merged, left_on='NBH_code', right_on='buurtcode', how = 'left')
cleaned_school = cleaned_school.loc[cleaned_school['buurtname'].isna() == False]
cleaned_school = cleaned_school[['buurtname','0-4age_Binnen5Km_55','4+age_Binnen10Km']]
cleaned_school.shape
```

```python
cleaned_school = cleaned_school.dropna(subset = ['0-4age_Binnen5Km_55','4+age_Binnen10Km'])
```

```python
cleaned_school.isna().sum()
#print(cleaned_school.dtypes)
```

```python
cleaned_school=cleaned_school.drop_duplicates(subset='buurtname', keep='first', inplace=False, ignore_index=False)
```
